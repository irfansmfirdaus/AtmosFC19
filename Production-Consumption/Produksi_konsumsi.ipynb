```python
from pylab import *
import matplotlib
import matplotlib.pyplot as plt
import xlrd
import numpy as np
from math import *
import pandas as pd
from datetime import datetime, timedelta
import pandas as pd
```

### Kab. Bogor

```python
konsumsi_bogor = np.array(sheet_data.col_values(1,99),dtype='float')
produksi_bogor = np.array(sheet_data.col_values(2,99))
```

```python
x= np.arange(len(produksi_bogor))
```

```python
beras_bogor = 0.6274*produksi_bogor
```

```python
fig, ax1 = plt.subplots(figsize = (12,4))

#ax1.set_ylabel('Curah hujan (mm)',fontsize=18)
ax1.set_title('BOGOR', fontsize=18)

ax1.plot(konsumsi_bogor,linestyle='--', marker='o', color='b', label= 'Rice Consumption (tons)')
ax1.plot(produksi_bogor,linestyle='--', marker='o', color='r',label= 'Paddy (tons)')
ax1.plot(beras_bogor,linestyle='--', marker='o', color='k',label= 'Rice (tons)')
plt.xticks(x, ('01/2018', '02/2018', '03/2018', '04/2018','05/2018','06/2018','07/2018','08/2018','09/2018','10/2018','11/2018','12/2018','01/2019', '02/2019', '03/2019', '04/2019','05/2019','06/2019','07/2019','08/2019','09/2019','10/2019','11/2019','12/2019'))
ax1.xaxis.set_tick_params(rotation=45)
ax1.set_ylim(0,250000)
ax1.tick_params(labelsize=12)
ax1.legend(loc=1, fontsize=12)
plt.savefig('produksi_konsumsi_bogor.png', bbox_inches='tight',dpi=300)
```

```python
surplus = beras_bogor-konsumsi_bogor
```

```python
fig, ax2 = plt.subplots(figsize = (12,4))

mask1 = surplus < 0
mask2 = surplus >= 0

ax2.set_title('BOGOR', fontsize=18)
plt.bar(x[mask1], surplus[mask1], color = 'red', label='Deficits (tons)')
#plt.bar(x[mask2], surplus[mask2], color = 'blue')
plt.xticks(x, ('01/2018', '02/2018', '03/2018', '04/2018','05/2018','06/2018','07/2018','08/2018','09/2018','10/2018','11/2018','12/2018','01/2019', '02/2019', '03/2019', '04/2019','05/2019','06/2019','07/2019','08/2019','09/2019','10/2019','11/2019','12/2019'))
ax2.xaxis.set_tick_params(rotation=45)
ax2.tick_params(labelsize=12)
ax2.set_ylim(-100000,100000)
#ax2.set_ylabel('Curah hujan (mm)',fontsize=18)
ax2.legend(loc=2, fontsize=12)
plt.savefig('surplus_defisit_bogor.png', bbox_inches='tight',dpi=300)
```
