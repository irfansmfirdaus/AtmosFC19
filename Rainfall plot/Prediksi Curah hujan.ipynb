```python
import numpy as np
from pylab import *
import matplotlib
import matplotlib.pyplot as plt
import xlrd
```

```python
data= xlrd.open_workbook('CH.xlsx')
sheet_data =data.sheet_by_index(0)
```

```python
dry_bandung = array(sheet_data.col_values(1,2))
norm_bandung = array(sheet_data.col_values(2,2))
wet_bandung = array(sheet_data.col_values(3,2))
```

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,1,1])
X= np.arange(12)

ax.set_ylabel('Curah hujan (mm)',fontsize=18)
#ax.set_xlabel('Bulan', fontsize=18)


ax.bar(X + -0.3, dry_bandung, color = 'r', width = 0.3, label= 'Dry years')
ax.bar(X + 0, norm_bandung, color = 'g', width = 0.3, label= 'Normal years')
ax.bar(X + 0.3, wet_bandung, color = 'b', width = 0.3, label= 'wet years')

ax.xaxis.set_tick_params(rotation=45)
ax.tick_params(labelsize=16)
ax.set_ylim(0,400)
ax.set_title('Kabupaten Bandung', fontsize=18)
plt.xticks(X, ('Januari', 'Februari', 'Maret', 'April','Mei','Juni','Juli','Agustus','September','Oktober','November','Desember'))
ax.legend(loc=2, fontsize=11)

plt.savefig('Kabbandung.png', bbox_inches='tight',dpi=500)
```
